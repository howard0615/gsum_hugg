```python
from transformers import BartForConditionalGeneration, BartModel
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large")
```

```python
model_static_dict = model.state_dict()
```

```python
for k in model_static_dict.keys():
    print(k)
```

```python
z_encoder = {}
for k, v in model_static_dict.items():
    if k.startswith("model.encoder."):
        z_encoder[k.replace("encoder", "z_encoder")]=v

z_encoder_attn = {}
for k, v in model_static_dict.items():
    if k.startswith("decoder.layers.") and "encoder_attn" in k:
        print(k)
        z_encoder_attn[k.replace("encoder_attn", "z_encoder_attn")]=v
```

```python
z_encoder.keys()
```

```python
from transformers import BartForConditionalGeneration

model = BartForConditionalGeneration.from_pretrained("fnlp/bart-large-chinese")
```

```python
model_static_dict = model.state_dict()

for k in model_static_dict.keys():
    print(k)
```

```python
z_encoder = {}
for k, v in model_static_dict.items():
    if k.startswith("model.encoder."):
        z_encoder[k.replace("encoder", "z_encoder")]=v

z_encoder_attn = {}
for k, v in model_static_dict.items():
    if k.startswith("model.decoder.layers.") and "encoder_attn" in k:
        print(k)
        z_encoder_attn[k.replace("encoder_attn", "z_encoder_attn")]=v
```

```python
z_encoder.keys()
```

```python
z_encoder_attn.keys()
```

```python
for k, v in z_encoder.items():
    model_static_dict[k] = v
for k, v in z_encoder_attn.items():
    model_static_dict[k] = v
```

```python
for k in model_static_dict.keys():
    print(k)
```

```python
import torch

torch.save(model_static_dict, "/workplace/yhcheng/gsum_like/chinese_gsum_like_bart/pytorch_model.bin")
```

```python
from transformers import BartForConditionalGeneration

model = BartForConditionalGeneration.from_pretrained("fnlp/bart-large-chinese")
```

```python
model_static_dict = model.state_dict()

for k in model_static_dict.keys():
    print(k)
```

```python
encoder_12 = {}
for k, v in model_static_dict.items():
    if k.startswith("model.encoder.layers.11"):
        encoder_12[k.replace("11", "12")]=v
```

```python
for k in encoder_12.keys():
    print(k)
```

```python
for k, v in encoder_12.items():
    model_static_dict[k]=v
```

```python
model_static_dict.keys()
```

```python
z_encoder_attn = {}
for k, v in model_static_dict.items():
    if k.startswith("model.decoder.layers.") and "encoder_attn" in k:
        print(k)
        z_encoder_attn[k.replace("encoder_attn", "z_encoder_attn")]=v
```

```python
for k, v in z_encoder_attn.items():
    model_static_dict[k] = v
```
